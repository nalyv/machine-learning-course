```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
X, y = load_boston(return_X_y= True)
boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 3)
```

```python
# Linear Model
linearModel = LinearRegression()
linearModel.fit(X_train, y_train)
linear_train_r2 = linearModel.score(X_train, y_train) # train R^2
linear_test_r2 = linearModel.score(X_test, y_test) # test R^2
print('Linear: Train R2 value: ', linear_train_r2)
print('Linear: Test R2 value: ', linear_test_r2)
```

**RIDGE**

```python
ridge = Ridge(alpha = 0.01) # lambda = 0.01
ridge.fit(X_train, y_train)
ridge_train_r2 = ridge.score(X_train, y_train)
ridge_test_r2 = ridge.score(X_test, y_test)
print('Ridge: Train R2 value: ', ridge_train_r2)
print('Ridge: Test R2 value: ', ridge_test_r2)
```

```python
ridge2 = Ridge(alpha = 100) # lambda = 100
ridge2.fit(X_train, y_train)
ridge2_train_r2 = ridge2.score(X_train, y_train)
ridge2_test_r2 = ridge2.score(X_test, y_test)
print('Ridge2: Train R2 value: ', ridge2_train_r2)
print('Ridge2: Test R2 value: ', ridge2_test_r2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(linearModel.coef_,
         alpha = 0.4,
         linestyle = 'none',
         marker = 'o',
         markersize = 15,
         color = 'green',
         label = 'Linear Model')
plt.plot(ridge.coef_, 
         alpha =0.7,
         linestyle = 'none',
         marker = '*',
         markersize = 15,
         color = 'red',
         label = r'Ridge: $\lambda = 0.01$')
plt.plot(ridge2.coef_,
         alpha = 0.5,
         linestyle = 'none',
         marker = 'd',
         markersize = 15,
         color = 'blue',
         label = r'Ridge2: $\lambda = 100$')
plt.xlabel('Coefficients', fontsize = 16)
plt.ylabel('Coefficient Values', fontsize = 16)
plt.legend(fontsize = 13, loc = 4)
```

**LASSO**

```python
lasso = Lasso(alpha = 0.01) # lambda = 0.01
lasso.fit(X_train, y_train)
lasso_train_r2 = lasso.score(X_train, y_train)
lasso_test_r2 = lasso.score(X_test, y_test)
print('Lasso: Train R^2 value', lasso_train_r2)
print('Lasso: Test R^2 value', lasso_test_r2)
```

```python
lasso2 = Lasso(alpha = 1) # lambda = 1
lasso2.fit(X_train, y_train)
lasso2_train_r2 = lasso2.score(X_train, y_train)
lasso2_test_r2 = lasso2.score(X_test, y_test)
print('Lasso2: Train R^2 value', lasso2_train_r2)
print('Lasso2: Test R^2 value', lasso2_test_r2)
```

```python
lasso3 = Lasso(alpha = 100) # lambda = 1
lasso3.fit(X_train, y_train)
lasso3_train_r2 = lasso3.score(X_train, y_train)
lasso3_test_r2 = lasso3.score(X_test, y_test)
print('Lasso3: Train R^2 value', lasso3_train_r2)
print('Lasso3: Test R^2 value', lasso3_test_r2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(linearModel.coef_,
         alpha = 0.7,
         linestyle = 'none',
         marker = 'v',
         markersize = 15,
         color = 'green',
         label = r'Linear Model$')
plt.plot(lasso.coef_,
         alpha = 0.7,
         linestyle = 'none',
         marker = '*',
         markersize = 15,
         color = 'red',
         label = r'Lasso: $\lambda = 0.01$')
plt.plot(lasso2.coef_,
         alpha = 0.7,
         linestyle = 'none',
         marker = 'd',
         markersize = 15,
         color = 'blue',
         label = r'Lasso2: $\lambda = 1$')
plt.plot(lasso3.coef_,
         alpha = 0.7,
         linestyle = 'none',
         marker = 's',
         markersize = 15,
         color = 'black',
         label = r'Lasso3: $\lambda = 100$')
plt.xlabel('Features', fontsize = 20)
plt.ylabel('Features Coefficients', fontsize = 20)
plt.legend(fontsize = 18, loc = 4)
```

```python
lasso_coef = sum(lasso.coef_ == 0)
lasso2_coef = sum(lasso2.coef_ == 0)
lasso3_coef = sum(lasso3.coef_ == 0)

print('Lasso (lambda = 0.01):', lasso_coef)
print('Lasso2 (lambda = 1):', lasso2_coef)
print('Lasso3 (lambda = 100):', lasso3_coef)
```
