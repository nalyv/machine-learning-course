```python
from sklearn.datasets import load_boston
import statsmodels.api as sm
import numpy as np
```

```python
boston = load_boston()
X = boston.data
y = boston.target
X = sm.add_constant(X)
features = list(boston.feature_names)
features.insert(0, 'Intercept')

# subset selection
for i in range(X.shape[1], 0, -1):
    linearModel = sm.regression.linear_model.OLS(y, X)
    fit = linearModel.fit()
    pvalues = fit.pvalues
    # finding the feature with the highest p value
    maxIndex = np.where(pvalues == max(pvalues))[0][0]
    print(features[maxIndex], '|', pvalues[maxIndex])
    X = np.delete(X, maxIndex, axis = 1)
    features.remove(features[maxIndex])
```
