```
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_wine
```

```
wine = load_wine()
X = wine.data
y = wine.target

cv = KFold(n_splits = 10)
tree = DecisionTreeClassifier(criterion = 'entropy')
```

```
bagging = BaggingClassifier(base_estimator = tree,
                            n_estimators = 5)
bagging_performance = cross_val_score(bagging, X, y, cv = cv)
print('Bagging performance: ', bagging_performance.mean())
```

```
random_forest = RandomForestClassifier(n_estimators = 5,
                                       max_features = 3)
random_forest_performance = cross_val_score(random_forest, X, y, cv = cv)
print('Random forest performance: ', random_forest_performance.mean())
```

```
adaboost = AdaBoostClassifier(base_estimator = tree, n_estimators = 1000)
adaboost_performance = cross_val_score(adaboost, X, y, cv = cv)
print('Adaboost performance: ', adaboost_performance.mean())
```
